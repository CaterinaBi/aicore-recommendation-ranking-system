```python
import pandas as pd
```

```python
products_df = pd.read_csv("products.csv", sep=",")
products_df.head(600)
```

```python
# assigns operation to dataframe variable
products_df = products_df.dropna() # deletes columns where at least 1 item is missing
```

```python
# cheks that NaN columns have been dropped
products_df.head(60)
```

```python
products_df.dtypes #finds the object type for each column
```

```python
products_df.info() # finds the object types and number of nulls
```

```python
# price column needs to be converted as a float
# if we try to sum objects, we don't get the expected result

# products_df['price'].sum() # throws error
```

```python
products_df['price'] = products_df['price'].str.strip('¬£')
products_df = products_df.replace(',','', regex=True) # commas need to go to convert price string to float
products_df['price'] = products_df['price'].astype('float64')
```
